```python
# initial imports
import pandas as pd
from path import Path
```

```python
# Load data
file_path = Path("../Resources/loans_data.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
# Use labelencoder to convert the categorical columns to numberical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Convert Month
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
df['month_num'] = df['month'].apply(lambda x: months_num[x])
df.head()
```

```python
# Convert Gender
gender_num = {
    'male': 0,
    'female': 1
}
df['gender_num'] = df['gender'].apply(lambda x: gender_num[x])
df.head()
```

```python
# Convert Education
education_num = {
    'High School or Below': 0,
    'college': 1,
    'Bachelor': 2,
    'Master or Above': 3
}
df['education_num'] = df['education'].apply(lambda x: education_num[x])
df.head()
```

```python
# Drop converted columns
df = df.drop(['month', 'education', 'gender'], axis=1)
df.head()
```

```python
# Now using LabelEncoder
df2 = pd.read_csv(file_path)
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
df2['month_num'] = df2['month'].apply(lambda x: months_num[x])
df2.head()
```

```python
# Gender and education
df2 = pd.get_dummies(df2, columns=["education", "gender"])
df2.head()
```
